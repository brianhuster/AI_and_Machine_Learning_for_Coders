## MÃ£ hÃ³a theo tá»« (Word Tokenizer) ğŸ“  

Hiá»ƒu Ä‘Æ¡n giáº£n lÃ  vá»›i má»—i tá»«, tá»¥i mÃ¬nh sáº½ gÃ¡n má»™t sá»‘ Ä‘á»ƒ Ä‘áº¡i diá»‡n cho tá»« Ä‘Ã³.

````python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.text import Tokenizer

org_texts = [
    "Today is a sunny day",
    "Today is a rainy day"
]


# Äá»‹nh nghÄ©a sá»‘ lÆ°á»£ng tá»« trong cÃ´ng cá»¥ mÃ£ hÃ³a (tokenizer)
tokenizer = Tokenizer(num_words = 100)
# Tiáº¿n hÃ nh Ä‘áº©y cÃ¡c tá»« vÃ o Ä‘á»ƒ táº¡o thÆ° viá»‡n tá»« trong cÃ´ng cá»¥ mÃ£ hÃ³a.
tokenizer.fit_on_texts(org_texts)
# Láº¥y tá»« Ä‘iá»ƒn mÃ£ hÃ³a ra
word_index = tokenizer.word_index
print(word_index)
````

## Using Out of Vocabulary Tokens (Sá»­ dá»¥ng cÃ¡c token cho tá»« vá»±ng má»›i) ğŸ†•ğŸ“š  


> **Váº­y chÃºng ta cÃ³ thá»ƒ giáº£i quyáº¿t váº¥n Ä‘á» "Out of Vocabulary" nÃ y nhÆ° tháº¿ nÃ o?** ğŸ¤”

á» Ä‘Ã¢y, mÃ¬nh chá»‰ giá»›i thiá»‡u cÃ¡ch Ä‘Æ¡n giáº£n nháº¥t thÃ´i nha. Má»i ngÆ°á»i sáº½ sá»­ dá»¥ng má»™t **token hay sá»‘** Ä‘á»ƒ Ä‘áº¡i diá»‡n cho cÃ¡c trÆ°á»ng há»£p **tá»« vá»±ng khÃ´ng biáº¿t trÆ°á»›c**. ğŸ†•ğŸ”¢  

á» Ä‘Ã¢y, mÃ¬nh sáº½ sá»­ dá»¥ng token lÃ  **< OOV >** nha. Má»i ngÆ°á»i cÅ©ng cÃ³ thá»ƒ Ä‘áº·t tÃªn token lÃ  cÃ¡i khÃ¡c tÃ¹y Ã½.  
Trong má»™t sá»‘ mÃ´ hÃ¬nh hoáº·c Ä‘oáº¡n code khÃ¡c, má»i ngÆ°á»i cÃ³ thá»ƒ báº¯t gáº·p token nÃ y dÆ°á»›i dáº¡ng **< UNK >**, cÃ³ nghÄ©a lÃ  **unknown - khÃ´ng biáº¿t**. â“

## Understanding padding

Hiá»ƒu Ä‘Æ¡n giáº£n, **Ä‘á»‡m chuá»—i** nghÄ©a lÃ  **thÃªm cÃ¡c token vÃ o chuá»—i**. ğŸ”¢  

NhÆ° má»i ngÆ°á»i Ä‘Ã£ biáº¿t, dá»¯ liá»‡u khi Ä‘Æ°a vÃ o mÃ´ hÃ¬nh há»c táº­p **pháº£i cÃ³ kÃ­ch thÆ°á»›c giá»‘ng nhau, tÆ°Æ¡ng á»©ng vá»›i kÃ­ch thÆ°á»›c Ä‘áº§u vÃ o cá»§a mÃ´ hÃ¬nh**. ğŸ› ï¸  

á» cÃ¡c chÆ°Æ¡ng trÆ°á»›c, khi chÃºng ta lÃ m viá»‡c vá»›i dá»¯ liá»‡u dáº¡ng áº£nh, chÃºng ta Ä‘Ã£ tiáº¿n hÃ nh Ä‘iá»u chá»‰nh láº¡i kÃ­ch thÆ°á»›c áº£nh trÆ°á»›c khi Ä‘Æ°a vÃ o mÃ´ hÃ¬nh. BÃ¢y giá», khi xá»­ lÃ½ dá»¯ liá»‡u dáº¡ng chuá»—i, quy trÃ¬nh cÅ©ng tÆ°Æ¡ng tá»± nhÆ° váº­y. ğŸ–¼ï¸â¡ï¸ğŸ”¡  

Vá»›i dá»¯ liá»‡u dáº¡ng chuá»—i, chÃºng ta sáº½ thÃªm vÃ o cÃ¡c token Ä‘á»ƒ Ä‘áº¡i diá»‡n cho nhá»¯ng pháº§n tá»­ Ä‘Æ°á»£c Ä‘á»‡m thÃªm vÃ o, vá»›i má»¥c Ä‘Ã­ch lÃ m cho Ä‘á»™ dÃ i cÃ¡c chuá»—i Ä‘á»u tÆ°Æ¡ng Ä‘Æ°Æ¡ng nhau. Token nÃ y thÆ°á»ng Ä‘Æ°á»£c tháº¥y dÆ°á»›i dáº¡ng **< PAD >** trong cÃ¡c mÃ´ hÃ¬nh xá»­ lÃ½. ğŸ§©  

NÃ³i Ä‘á»§ nhiá»u rá»“i, bÃ¢y giá» tá»¥i mÃ¬nh sáº½ Ä‘i vÃ o code xá»­ lÃ½ cá»¥ thá»ƒ nha. ğŸ’»âœ¨  


````python
# CÃ¡c cÃ¢u vÄƒn cáº§n mÃ£ hÃ³a Ä‘á»ƒ Ä‘Æ°a vÃ o mÃ´ hÃ¬nh
sentences = [
  "Today is a sunny day",
  "Today is a rainy day",
  "Is it sunny today?",
  "I really enjoyed walking in the snow today"
]

# Äá»‹nh nghÄ©a bá»™ mÃ£ hÃ³a
tokenizer = Tokenizer(num_words=100, oov_token="<OOV>")
tokenizer.fit_on_texts(sentences)
word_index = tokenizer.word_index

# Tiáº¿n hÃ nh mÃ£ hÃ³a cÃ¡c cÃ¢u vÄƒn
sequences = tokenizer.texts_to_sequences(sentences)
````

````python
# In ra káº¿t quáº£ mÃ£ hÃ³a
for seq in sequences:
  print(seq)
````

NhÆ° má»i  ngÆ°á»i tháº¥y Ä‘áº¥y, chÃºng cÃ³ kÃ­ch thÆ°á»›c khÃ´ng giá»‘ng nhau nha.

````python
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Tiáº¿n hÃ nh Ä‘á»‡m cÃ¡c chuá»—i
padded = pad_sequences(sequences)

# In ra káº¿t quáº£ Ä‘á»‡m
print("Káº¿t quáº£ cÃ¡c cÃ¢u Ä‘Ã£ Ä‘á»‡m:")
for seq in padded:
  print(seq)
````

CÃ¡c sá»‘ 0 sáº½ Ä‘áº¡i diá»‡n cho token **< PAD >** Ä‘Æ°á»£c Ä‘á»‡m á»Ÿ Ä‘áº§u. ÄÃ¢y lÃ  lÃ½ do mÃ  cÃ¡c bá»™ mÃ£ hÃ³a thÆ°á»ng cÃ³ tá»« Ä‘iá»ƒn báº¯t Ä‘áº§u tá»« sá»‘ 1, cÃ²n sá»‘ 0 sáº½ Ä‘Æ°á»£c dÃ nh cho pháº§n **padding** nha. ğŸ§©  

Viá»‡c thÃªm bá»™ Ä‘á»‡m tá»©c cÃ¡c sá»‘ 0 á»Ÿ pháº§n Ä‘áº§u chuá»—i nÃ y Ä‘Æ°á»£c gá»i lÃ  **prepadding**. ğŸ•’  

Náº¿u má»i ngÆ°á»i muá»‘n thÃªm bá»™ Ä‘á»‡m á»Ÿ phÃ­a sau, cÃ³ thá»ƒ sá»­ dá»¥ng thÃªm má»™t tham sá»‘ nhá» lÃ  **padding = 'post'** cho phÆ°Æ¡ng thá»©c trÃªn. ğŸ› ï¸âœ¨  


````python
padded = pad_sequences(sequences, padding='post')
print("Káº¿t quáº£ cÃ¡c cÃ¢u Ä‘Ã£ Ä‘á»‡m:")
for seq in padded:
  print(seq)
````

Äáº¥y, cÃ¡c sá»‘ 0 Ä‘Ã£ Ä‘Æ°á»£c Ä‘á»‡m vÃ o pháº§n Ä‘uÃ´i rá»“i nÃ y, qua Ä‘Ã³ giÃºp Ä‘áº£m báº£o kÃ­ch thÆ°á»›c cÃ¡c batch trÆ°á»›c khi Ä‘Æ°a vÃ o huáº¥n luyá»‡n mÃ´ hÃ¬nh. ğŸ”¢

Äá»ƒ kháº¯c phá»¥c tÃ¬nh tráº¡ng nÃ y, ta cÃ³ thá»ƒ Ä‘á»‹nh nghÄ©a má»™t tham sá»‘ gá»i lÃ  **Ä‘á»™ dÃ i tá»‘i Ä‘a (maxlen)** Ä‘á»ƒ giá»›i háº¡n Ä‘á»™ dÃ i chuá»—i. ğŸ”§

Trong khi Ä‘Ã³, chuá»—i cuá»‘i cÃ¹ng láº¡i bá»‹ cáº¯t ngáº¯n chá»‰ cÃ²n 6 kÃ½ tá»± cuá»‘i. ÄÃ¢y cÅ©ng lÃ  **má»™t sá»± Ä‘Ã¡nh Ä‘á»•i.** âš–ï¸

- **Loáº¡i bá» tá»« dá»«ng** (stopwords). âŒğŸ“

*(LÆ°u Ã½: danh sÃ¡ch nÃ y cÃ³ thá»ƒ chÆ°a Ä‘áº§y Ä‘á»§ táº¥t cáº£ cÃ¡c tá»« dá»«ng.)* âš ï¸

 **Äáº§u tiÃªn**: Loáº¡i bá» cÃ¡c tháº» HTML (**HTML tags**). Má»i ngÆ°á»i cÃ³ thá»ƒ sá»­ dá»¥ng thÆ° viá»‡n **BeautifulSoup** Ä‘á»ƒ thá»±c hiá»‡n dá»… dÃ ng hÆ¡n. ğŸ“œ  

DÆ°á»›i Ä‘Ã¢y lÃ  vÃ­ dá»¥ vá» cÃ¡ch sá»­ dá»¥ng: ğŸ’»


````python
# Import thÆ° viá»‡n vÃ o
from bs4 import BeautifulSoup
````

````python
# Táº¡o cÃ¢u máº«u cÃ³ chá»©a tháº» HTML trong Ä‘Ã³
sentence = "Captain Teemo <br> on duty!" # <br> lÃ  tháº» trong HTML Ä‘á»ƒ xuá»‘ng hÃ ng
# Cho cÃ¢u máº«u qua BeautifulSoup Ä‘á»ƒ táº¡o Ä‘á»‘i tÆ°á»£ng soup Ä‘Ã£ Ä‘Æ°á»£c xá»­ lÃ½
soup = BeautifulSoup(sentence)
# Tiáº¿n hÃ nh chá»‰ trÃ­ch xuáº¥t má»—i dá»¯ liá»‡u vÄƒn báº£n, khÃ´ng bao gá»“m cÃ¡c tháº» HTML
sentence = soup.get_text()
print("CÃ¢u Ä‘Ã£ Ä‘Æ°á»£c loáº¡i bá» tháº» HTML: ", sentence)
````

Má»i ngÆ°á»i tháº¥y Ä‘áº¥y, tháº» **< br >** Ä‘Ã£ bá»‹ loáº¡i bá», hiá»‡n táº¡i khoáº£ng tráº¯ng giá»¯a 2 tá»« **Teemo** vÃ  **on** lÃ  2 kÃ½ tá»± khoáº£ng cÃ¡ch (space).


BÃ¢y giá» tá»¥i mÃ¬nh sáº½ tiáº¿n hÃ nh loáº¡i bá» tá»« dá»«ng cÅ©ng nhÆ° cÃ¡c dáº¥u cÃ¢u nha. ğŸ§¹  

á» Ä‘Ã¢y, cÃ¡ch Ä‘Æ¡n giáº£n nháº¥t lÃ  má»i ngÆ°á»i láº·p qua cÃ¡c tá»« vÃ  kiá»ƒm tra xem tá»« Ä‘Ã³ cÃ³ trong danh sÃ¡ch tá»« dá»«ng khÃ´ng nha, náº¿u cÃ³ thÃ¬ tiáº¿n hÃ nh loáº¡i bá», cÃ²n khÃ´ng thÃ¬ giá»¯ láº¡i. âŒâœ…  

Vá» pháº§n cÃ¡c dáº¥u cÃ¢u, chÃºng ta cÃ³ thá»ƒ sá»­ dá»¥ng phÆ°Æ¡ng thá»©c **translate** Ä‘á»ƒ tiáº¿n hÃ nh loáº¡i bá». Má»i ngÆ°á»i cÃ³ thá»ƒ dÃ¹ng **string.punctuation** Ä‘á»ƒ láº¥y danh sÃ¡ch cÃ¡c kÃ½ tá»± Ä‘áº·c biá»‡t vÃ  dáº¥u cÃ¢u nha. âœ‚ï¸ğŸ”¡  


````python
import string
# In ra danh sÃ¡ch cÃ¡c dáº¥u cÃ¢u vÃ  kÃ­ tá»± Ä‘áº·c biá»‡t.
print(string.punctuation)
````

````python
# Táº¡o báº£ng mapping Ä‘á»ƒ loáº¡i bá» kÃ­ tá»± Ä‘áº·c biá»‡t
table = str.maketrans('', '', string.punctuation) # Chuyá»ƒn cÃ¡c kÃ­ tá»± Ä‘áº·c biá»‡t thÃ nh kÃ­ tá»± rá»—ng
````

````python
# Láº¥y danh sÃ¡ch tá»« dá»«ng, á»Ÿ Ä‘Ã¢y mÃ¬nh sá»­ dá»¥ng thÆ° viá»‡n nltk Ä‘á»ƒ láº¥y tá»« dá»«ng trong tiáº¿ng Anh nha
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
stopwords = set(stopwords.words('english'))
print("Danh sÃ¡ch tá»« dá»«ng: ", stopwords)
````

````python
# Chuyá»ƒn danh sÃ¡ch tá»« dá»«ng sang dáº¡ng list
stopwords = list(stopwords)
````

````python
# Chia thÃ nh danh sÃ¡ch cÃ¡c tá»« ra
words = sentence.split()
print("Danh sÃ¡ch cÃ¡c tá»«: ", words)
# Chuá»—i káº¿t quáº£
filtered_sentence = ""
for word in words:
  word = word.translate(table)    # Loáº¡i bá» cÃ¡c kÃ­ tá»± Ä‘áº·c biá»‡t
  if word not in stopwords:       # Kiá»ƒm tra cÃ³ thuá»™c tá»« dá»«ng khÃ´ng Ä‘á»ƒ bá» qua.
    filtered_sentence = filtered_sentence + word + " "
print("CÃ¢u Ä‘Ã£ Ä‘Æ°á»£c tiá»n xá»­ lÃ½: ", filtered_sentence)
````

# Working with Real Data Sources (LÃ m viá»‡c vá»›i cÃ¡c nguá»“n dá»¯ liá»‡u thá»±c táº¿) ğŸ“ŠğŸŒ  


Tá»¥i mÃ¬nh sáº½ sá»­ dá»¥ng **TFDS** Ä‘Ã£ há»c á»Ÿ **chÆ°Æ¡ng 4** Ä‘á»ƒ táº£i dá»¯ liá»‡u nha. ğŸ“¦  

Bá»™ dá»¯ liá»‡u láº§n nÃ y tá»¥i mÃ¬nh dÃ¹ng lÃ  **imdb_reviews**, gá»“m 50.000 bÃ¬nh luáº­n Ä‘Ã¡nh giÃ¡ phim, Ä‘Æ°á»£c gáº¯n nhÃ£n lÃ  tÃ­ch cá»±c hoáº·c tiÃªu cá»±c nha. ğŸ¥ğŸ‘ğŸ‘  


````python
import tensorflow as tf
import tensorflow_datasets as tfds
````

Tá»¥i mÃ¬nh sáº½ tiáº¿n hÃ nh láº·p qua bá»™ dá»¯ liá»‡u vÃ  thÃªm pháº§n vÄƒn báº£n bÃ¬nh luáº­n vÃ o má»™t danh sÃ¡ch Ä‘á»ƒ dá»… xá»­ lÃ½ nha. ğŸ“  

Tá»¥i mÃ¬nh sá»­ dá»¥ng **tfds.as_numpy** Ä‘á»ƒ Ä‘áº£m báº£o ráº±ng dá»¯ liá»‡u tráº£ vá» sáº½ á»Ÿ dáº¡ng **string** chá»© khÃ´ng pháº£i **tensor**. ğŸ”§  


````python
train_data = tfds.as_numpy(tfds.load('imdb_reviews', split='train'))
````

````python
# Kiá»ƒm tra dá»¯ liá»‡u cá»§a máº«u (item) Ä‘áº§u tiÃªn bÃªn trong bá»™ dá»¯ liá»‡u
item_1st = next(iter(train_data))
print(item_1st)
print(type(item_1st))
````

Váº­y lÃ  má»—i máº«u (item) Ä‘Æ°á»£c lÆ°u vá»›i dáº¡ng má»™t **dictionary** vá»›i 2 khÃ³a lÃ  **'text' vÃ  'label'**

````python
imdb_sentences = []
# Tiáº¿n hÃ nh láº·p qua tá»«ng máº«u (item) Ä‘á»ƒ trÃ­ch xuáº¥t vÄƒn báº£n ra lÆ°u vÃ o danh sÃ¡ch imdb_sentences
for item in train_data:
  imdb_sentences.append(str(item['text']))
````

````python
# Xem thá»­ 5 bÃ¬nh luáº­n Ä‘Ã¡nh giÃ¡ phim Ä‘áº§u tiÃªn
for idx, comment in enumerate(imdb_sentences[:5]):
  print(f"BÃ¬nh luáº­n {idx+1}: {comment}")
  print("-"*100)
````

BÃ¢y giá» tá»¥i mÃ¬nh tiáº¿n hÃ nh táº¡o bá»™ mÃ£ hÃ³a cho bá»™ dá»¯ liá»‡u nha.

````python
from tensorflow.keras.preprocessing.text import Tokenizer
# Khá»Ÿi táº¡o bá»™ mÃ£ hÃ³a
tokenizer = Tokenizer(num_words=5000)
# Tiáº¿n hÃ nh Ä‘Æ°a dá»¯ liá»‡u vÃ o bá»™ mÃ£ hÃ³a, táº¡o tá»« Ä‘iá»ƒn
tokenizer.fit_on_texts(imdb_sentences)
# Tiáº¿n hÃ nh mÃ£ hÃ³a cÃ¡c bÃ¬nh luáº­n
sequences = tokenizer.texts_to_sequences(imdb_sentences)
````

```

```



````python
# Táº£i danh sÃ¡ch tá»« dá»«ng
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
stopwords = set(stopwords.words('english'))

from bs4 import BeautifulSoup
import string

table = str.maketrans('', '', string.punctuation)

# Táº£i bá»™ dá»¯ liá»‡u vÃ o danh sÃ¡ch bÃ¬nh luáº­n
imdb_sentences = []
train_data = tfds.as_numpy(tfds.load('imdb_reviews', split='train'))
for item in train_data:
  sentence = str(item['text'].decode('UTF-8').lower()) # Tiáº¿n hÃ nh chuyá»ƒn sang dáº¡ng UTF-8 vÃ  dáº¡ng chá»¯ thÆ°á»ng
  soup = BeautifulSoup(sentence)
  sentence = soup.get_text() # Loáº¡i bá» cÃ¡c tháº» HTML
  words = sentence.split()
  filtered_sentence = ""
  for word in words:
    word = word.translate(table) # Loáº¡i bá» kÃ­ tá»± Ä‘áº·c biá»‡t
    if word not in stopwords: # Loáº¡i bá» tá»« dá»«ng
      filtered_sentence = filtered_sentence + word + " "
  imdb_sentences.append(filtered_sentence)

# Sau khi tiáº¿n hÃ nh lÃ m sáº¡ch dá»¯ liá»‡u, chÃºng ta sáº½ tiáº¿n hÃ nh mÃ£ hÃ³a
tokenizer = Tokenizer(num_words=25000)
tokenizer.fit_on_texts(imdb_sentences)
sequences = tokenizer.texts_to_sequences(imdb_sentences)  # MÃ£ hÃ³a dá»¯ liá»‡u
````

````python
import nltk
from nltk.corpus import stopwords

# Chuáº©n bá»‹ trÆ°á»›c cho pháº§n loáº¡i bá» tá»« dá»«ng
nltk.download('stopwords')
stopwords = set(stopwords.words('english'))

# Chuáº©n bá»‹ trÆ°á»›c cho pháº§n loáº¡i bá» tháº» HTML
from bs4 import BeautifulSoup

# Chuáº©n bá»‹ trÆ°á»›c cho pháº§n loáº¡i bá» kÃ­ tá»± Ä‘áº·c biá»‡t
import string
table = str.maketrans('', '', string.punctuation)
````

````python
sentences = []
labels = []
for r in rows[1:]: # VÃ¬ hÃ ng Ä‘áº§u tiÃªn lÃ  tÃªn cÃ¡c cá»™t nÃªn chÃºng ta khÃ´ng láº¥y, tÃ­nh tá»« vá»‹ trÃ­ index 1 nha
  labels.append(r[1])
  sentence = r[2].lower()
  # Tiáº¿n hÃ nh tiá»n xá»­ lÃ½ cÃ¡c kÃ­ tá»± Ä‘áº·c biá»‡t
  sentence = sentence.replace("/", " / ")
  sentence = sentence.replace("'", " ' ")
  sentence = sentence.replace("-", " - ")
  sentence = sentence.replace(".", " . ")
  # Tiáº¿n hÃ nh loáº¡i bá» cÃ¡c tháº» HTML báº±ng BeautifulSoup
  soup = BeautifulSoup(sentence)
  sentence = soup.get_text()
  # Chia thanh danh sÃ¡ch cÃ¡c tá»«
  words = sentence.split()
  filtered_sentence = ""
  # Tiáº¿n hÃ nh loáº¡i bá» kÃ­ tá»± Ä‘áº·c biá»‡t vÃ  kiá»ƒm tra Ä‘á»ƒ loáº¡i bá» cÃ¡c tá»« dá»«ng
  for word in words:
    word = word.translate(table) # Loáº¡i bá» kÃ­ tá»± Ä‘áº·c biá»‡t
    if word not in stopwords: # Kiá»ƒm tra vÃ  loáº¡i bá» tá»« dá»«ng
      filtered_sentence = filtered_sentence + word + " "
  sentences.append(filtered_sentence)
````
