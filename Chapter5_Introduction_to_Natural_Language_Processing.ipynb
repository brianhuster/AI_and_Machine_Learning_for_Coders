## MÃ£ hÃ³a theo tá»« (Word tokenizer)
Hiá»ƒu Ä‘Æ¡n giáº£n lÃ  vá»›i má»—i tá»« tá»¥i mÃ¬nh sáº½ cÃ³ má»™t sá»‘ Ä‘á»ƒ Ä‘áº¡i diá»‡n cho tá»« Ä‘Ã³.

```python
# "Caption Teemo on duty" -> [1, 2, 3, 4]
```

Tuy nhiÃªn chÃºng cÃ³ sáº½ cÃ³ cÃ¡c háº¡n cháº¿ khÃ¡c, tá»¥i mÃ¬nh sáº½ bÃ n sau ha. CÃ²n bÃ¢y giá» Ä‘i vÃ o vÃ­ dá»¥ code cá»¥ thá»ƒ trÆ°á»›c nha.

````python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.text import Tokenizer

org_texts = [
    "Today is a sunny day",
    "Today is a rainy day"
]


# Äá»‹nh nghÄ©a sá»‘ lÆ°á»£ng tá»« trong cÃ´ng cá»¥ mÃ£ hÃ³a (tokenizer)
tokenizer = Tokenizer(num_words = 100)
# Tiáº¿n hÃ nh Ä‘áº©y cÃ¡c tá»« vÃ o Ä‘á»ƒ táº¡o thÆ° viá»‡n tá»« trong cÃ´ng cá»¥ mÃ£ hÃ³a.
tokenizer.fit_on_texts(org_texts)
# Láº¥y tá»« Ä‘iá»ƒn mÃ£ hÃ³a ra
word_index = tokenizer.word_index
print(word_index)
````

NhÆ° Ä‘Ã¢y lÃ  káº¿t quáº£ cá»§a Ä‘oáº¡n code á»Ÿ trÃªn nha, chÃºng ta sáº½ láº¥y Ä‘Æ°á»£c tá»« Ä‘iá»ƒn mÃ£ hÃ³a tá»« nhÆ° sau

Sau khi Ä‘áº©y cÃ¡c chuá»—i vÃ o Ä‘á»ƒ táº¡o tá»« Ä‘iá»ƒn, tá»¥i mÃ¬nh sáº½ tiáº¿n mÃ£ hÃ³a cÃ¡c chuá»—i hay cÃ¢u vá»›i hÃ m **texts_to_sequences** nha. Má»i ngÆ°á»i chá»‰ cáº§n chuyá»ƒn list cÃ¡c cÃ¢u vÃ o.


````python
sequences = tokenizer.texts_to_sequences(org_texts)
print(sequences)
````

> **Váº­y Ä‘iá»ƒm háº¡n cháº¿ cá»§a phÆ°Æ¡ng phÃ¡p mÃ£ hÃ³a theo tá»« lÃ  gÃ¬?**

Váº­y, khi gáº·p nhá»¯ng **tá»« má»›i** khÃ´ng cÃ³ trong thÆ° viá»‡n tá»« vá»±ng, mÃ¡y tÃ­nh sáº½ khÃ´ng thá»ƒ phÃ¡t hiá»‡n Ä‘Æ°á»£c dáº«n Ä‘áº¿n chÃºng **bá» qua** nÃ³, lÃºc nÃ y khi ta cho mÃ´ hÃ¬nh há»c chÃºng sáº½ khÃ´ng Ä‘Æ°á»£c há»c Ä‘áº§y Ä‘á»§ vÃ  cÃ³ thá»ƒ bá»‹ nháº§m láº«n ngá»¯ cáº£nh.

VÃ­ dá»¥ nhÆ° Ä‘oáº¡n code dÆ°á»›i Ä‘Ã¢y.

````python
# Táº­p kiá»ƒm tra
test_data = [
    "Today is a snowy day",
    "Will it be rainy tomorrow?"
]

# Tiáº¿n hÃ nh mÃ£ hÃ³a
test_seq = tokenizer.texts_to_sequences(test_data)

# Kiá»ƒm tra káº¿t quáº£ mÃ£ hÃ³a.
print(test_seq)
````

ChÃºng ta cÃ³ tÃªn gá»i cá»¥ thá»ƒ cá»§a trÆ°á»ng há»£p nÃ y lÃ   **Out of vocabulary**, tá»©c náº±m ngoÃ i vÃ¹ng tá»« vá»±ng.

## Using Out of vocabulary tokens (Sá»­ dá»¥ng cÃ¡c token cho tá»« vá»±ng má»›i)

> Váº­y chÃºng ta cÃ³ thá»ƒ giáº£i quyáº¿t váº¥n Ä‘á» **Out of vocabulary** nÃ y nhÆ° tháº¿ nÃ o?

CÃ³ nhiá»u cÃ¡ch mÃ£ hÃ³a cáº£i tiáº¿n Ä‘á»ƒ kháº¯c phá»¥c hoáº·c háº¡n cháº¿ áº£nh hÆ°á»Ÿng cá»§a tÃ¬nh tráº¡ng nÃ y. á» Ä‘Ã¢y mÃ¬nh chá»‰ giá»›i thiá»‡u cÃ¡ch Ä‘Æ¡n giáº£n nháº¥t thÃ´i ha.
Má»i ngÆ°á»i sáº½ sá»­ dá»¥ng má»™t **token hay sá»‘** Ä‘á»ƒ Ä‘áº¡i diá»‡n cho cÃ¡c trÆ°á»ng há»£p **tá»« vá»±ng khÃ´ng biáº¿t trÆ°á»›c**.

á» Ä‘Ã¢y mÃ¬nh sáº½ sá»­ dá»¥ng token lÃ  **< OOV >** nha. Má»i ngÆ°á»i cÃ³ thá»ƒ Ä‘áº·t tÃªn token lÃ  cÃ¡i khÃ¡c cÅ©ng Ä‘Æ°á»£c. Trong má»™t sá»‘ mÃ´ hÃ¬nh hay Ä‘oáº¡n code má»i ngÆ°á»i cÃ³ thá»ƒ báº¯t gáº·p nÃ³ dÆ°á»›i dáº¡ng **< UNK >** cÃ³ nghÄ©a lÃ  **unknown - khÃ´ng biáº¿t**.

````python
# Tiáº¿n hÃ nh Ä‘á»‹nh nghÄ©a láº¡i bá»™ mÃ£ hÃ³a, thÃªm vÃ o token Ä‘áº¡i diá»‡n cho trÆ°á»ng há»£p tá»« khÃ´ng biáº¿t
tokenizer = Tokenizer(num_words=100, oov_token="<OOV>")
# Tiáº¿n hÃ nh Ä‘áº©y cÃ¡c tá»« vÃ o Ä‘á»ƒ táº¡o thÆ° viá»‡n tá»« trong cÃ´ng cá»¥ mÃ£ hÃ³a.
tokenizer.fit_on_texts(org_texts)
````

````python
# In tá»« Ä‘iá»ƒn bÃªn trong bá»™ mÃ£ hÃ³a ra
word_index = tokenizer.word_index
print(word_index)
````

````python
# Tiáº¿n hÃ nh mÃ£ hÃ³a láº¡i cÃ¡c cÃ¢u gáº·p pháº£i trÆ°á»ng há»£p tá»« vá»±ng khÃ´ng biáº¿t trÆ°á»›c Ä‘Ã³
test_seq = tokenizer.texts_to_sequences(test_data)
print(test_seq)
````

## Understanding padding

Hiá»ƒu Ä‘Æ¡n giáº£n thÃ¬ lÃ  **Ä‘á»‡m thÃªm cÃ¡c token vÃ o chuá»—i**. NhÆ° má»i ngÆ°á»i Ä‘Ã£ biáº¿t dá»¯ liá»‡u khi Ä‘Æ°a vÃ o mÃ´ hÃ¬nh há»c táº­p thÃ¬ **pháº£i cÃ³ kÃ­ch thÆ°á»›c giá»‘ng nhau, tÆ°Æ¡ng á»©ng vá»›i kÃ­ch thÆ°á»›c Ä‘áº§u vÃ o cá»§a mÃ´ hÃ¬nh**.

á» cÃ¡c chÆ°Æ¡ng trÆ°á»›c, khi chÃºng ta lÃ m viá»‡c vá»›i dá»¯ liá»‡u dáº¡ng áº£nh, chÃºng ta Ä‘Ã£ tiáº¿n hÃ nh Ä‘iá»u chá»‰nh láº¡i kÃ­ch thÆ°á»›c áº£nh trÆ°á»›c khi Ä‘Æ°a vÃ o mÃ´ hÃ¬nh thÃ¬ bÃ¢y giá» khi xá»­ lÃ½ dá»¯ liá»‡u dáº¡ng chuá»—i cÅ©ng tÆ°Æ¡ng tá»± nhÆ° váº­y Ã¡.

Vá»›i dá»¯ liá»‡u dáº¡ng chuá»—i chÃºng ta sáº½ thÃªm vÃ o cÃ¡c token Ä‘áº¡i diá»‡n cho nhá»¯ng pháº§n tá»­ Ä‘á»‡m thÃªm vÃ o vá»›i má»¥c Ä‘Ã­ch lÃ m cho Ä‘á»™ dÃ i cÃ¡c chuá»—i Ä‘á»u tÆ°Æ¡ng Ä‘Æ°Æ¡ng nhau. Má»i ngÆ°á»i sáº½ thÆ°á»ng báº¯t gáº·p token nÃ y vá»›i dáº¡ng **< PAD >>** trong cÃ¡c mÃ´ hÃ¬nh xá»­ lÃ½.

NÃ³i Ä‘á»§ nhiá»u rá»“i, bÃ¢y giá» chÃºng ta sáº½ Ä‘i vÃ o code xá»­ lÃ½ cá»¥ thá»ƒ ha.

````python
# CÃ¡c cÃ¢u vÄƒn cáº§n mÃ£ hÃ³a Ä‘á»ƒ Ä‘Æ°a vÃ o mÃ´ hÃ¬nh
sentences = [
  "Today is a sunny day",
  "Today is a rainy day",
  "Is it sunny today?",
  "I really enjoyed walking in the snow today"
]

# Äá»‹nh nghÄ©a bá»™ mÃ£ hÃ³a
tokenizer = Tokenizer(num_words=100, oov_token="<OOV>")
tokenizer.fit_on_texts(sentences)
word_index = tokenizer.word_index

# Tiáº¿n hÃ nh mÃ£ hÃ³a cÃ¡c cÃ¢u vÄƒn
sequences = tokenizer.texts_to_sequences(sentences)
````

````python
# In ra káº¿t quáº£ mÃ£ hÃ³a
for seq in sequences:
  print(seq)
````

NhÆ° má»i  ngÆ°á»i tháº¥y Ä‘áº¥y, chÃºng cÃ³ kÃ­ch thÆ°á»›c khÃ´ng giá»‘ng nhau nha.


```python

[2, 3, 4, 7, 6]
[3, 8, 5, 2]
[9, 10, 11, 12, 13, 14, 15, 2]
```
BÃ¢y giá» tá»¥i mÃ¬nh sáº½ tiáº¿n hÃ nh Ä‘á»‡m vÃ o Ä‘á»ƒ giÃºp Ä‘á»™ dÃ i cÃ¡c chuá»—i cÃ³ cÃ¹ng kÃ­ch thÆ°á»›c vá»›i chuá»—i cÃ³ Ä‘á»™ dÃ i nháº¥t.


````python
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Tiáº¿n hÃ nh Ä‘á»‡m cÃ¡c chuá»—i
padded = pad_sequences(sequences)

# In ra káº¿t quáº£ Ä‘á»‡m
print("Káº¿t quáº£ cÃ¡c cÃ¢u Ä‘Ã£ Ä‘á»‡m:")
for seq in padded:
  print(seq)
````

NhÆ° má»i ngÆ°á»i tháº¥y, bÃ¢y giá» cÃ¡c chuá»—i Ä‘á»u Ä‘Ã£ cÃ³ kÃ­ch thÆ°á»›c lÃ  8 ha. Tuy nhiÃªn cÃ¡c sá»‘ 0 Ä‘áº¡i diá»‡n token **< PAD >** láº¡i Ä‘Æ°á»£c Ä‘á»‡m á»Ÿ Ä‘áº§u. ÄÃ¢y cÅ©ng lÃ  lÃ½ do mÃ  thÆ°á»ng cÃ¡c bá»™ mÃ£ hÃ³a cÃ³ tá»« Ä‘iá»ƒn báº¯t Ä‘áº§u tá»« 1 ha, sá»‘ 0 sáº½ dÃ nh cho pháº§n **padding**.

Viá»‡c thÃªm bá»™ Ä‘á»‡m á»Ÿ pháº§n Ä‘áº§u nÃ y tá»¥i mÃ¬nh sáº½ gá»i lÃ  **prepadding**.

Náº¿u má»i ngÆ°á»i muá»‘n thÃªm bá»™ Ä‘á»‡m á»Ÿ phÃ­a sau, má»i ngÆ°á»i cÃ³ thá»ƒ dÃ¹ng thÃªm má»™t tham sá»‘ nhá» lÃ  **padding = 'post'** cho phÆ°Æ¡ng thá»©c trÃªn.

````python
padded = pad_sequences(sequences, padding='post')
print("Káº¿t quáº£ cÃ¡c cÃ¢u Ä‘Ã£ Ä‘á»‡m:")
for seq in padded:
  print(seq)
````

Äáº¥y, cÃ¡c sá»‘ 0 Ä‘Ã£ Ä‘Æ°á»£c Ä‘á»‡m vÃ o pháº§n Ä‘uÃ´i rá»“i nÃ y, qua Ä‘Ã³ giÃºp Ä‘áº£m báº£o kÃ­ch thÆ°á»›c cÃ¡c batch trÆ°á»›c khi Ä‘Æ°a vÃ o huáº¥n luyá»‡n mÃ´ hÃ¬nh.

ThÃ´ng qua viá»‡c thÃªm cÃ¡c bá»™ Ä‘á»‡m vÃ o, thÃ¬ tá»¥i mÃ¬nh Ä‘Ã£ cÃ³ thá»ƒ Ä‘áº£m báº£o Ä‘Æ°Æ¡c Ä‘á»™ dÃ i cÃ¡c thÃ nh pháº§n trong batch. Tuy nhiÃªn báº¡n cÅ©ng sáº½ Ä‘á»‘i máº·t vá»›i má»™t trÆ°á»ng há»£p Ä‘áº·c biá»‡t.

Trong khi Ä‘Ã³ chuá»—i cuá»‘i cÃ¹ng láº¡i Ä‘Æ°á»£c cáº¯t ngáº¯n chá»‰ cÃ²n 6 kÃ­ tá»± cuá»‘i. ÄÃ¢y cÅ©ng lÃ  **má»™t sá»± Ä‘Ã¡nh Ä‘á»•i**. ChÃºng ta cháº¥p nháº­n loáº¡i bá» Ä‘i má»™t sá»‘ dá»¯ liá»‡u hay tá»« nháº¥t Ä‘á»‹nh tá»« má»™t máº«u Ä‘á»ƒ giá»¯ láº¡i sá»± há»¯u Ã­ch tá»« cÃ¡c máº«u khÃ¡c.

Äiá»u nÃ y Ä‘á»“ng nghÄ©a vá»›i viá»‡c dá»¯ liá»‡u sáº½ chá»©a nhiá»u nhiá»…u hÆ¡n cÅ©ng nhÆ° cáº§n pháº£i tráº£i qua cÃ¡c bÆ°á»›c tiá»n xá»­ lÃ½ vÃ  lÃ m sáº¡ch mÃ  Ä‘iá»ƒn hÃ¬nh lÃ  **loáº¡i bá» tá»« dá»«ng** cÅ©ng nhÆ° cÃ¡c **dáº¥u cÃ¢u, kÃ­ tá»± Ä‘áº·c biá»‡t**.

Dá»¯ liá»‡u thu tháº­p Ä‘Æ°á»£c trong quÃ¡ trÃ¬nh cÃ o tá»« cÃ¡c web khÃ´ng pháº£i lÃºc nÃ o cÅ©ng hoÃ n háº£o cáº£, Ä‘Ã´i khi sáº½ chá»©a cÃ¡c **tháº» HTML** trong Ä‘Ã³ hay cÃ¡c tá»« thÃ´ lá»—, dáº¥u cÃ¢u hoáº·c tÃªn riÃªng. Do Ä‘Ã³ ta sáº½ tiáº¿n hÃ nh xá»­ lÃ½ cÃ¡c pháº§n nÃ y.

Äáº§u tiÃªn lÃ  loáº¡i bá» cÃ¡c tháº» HTML (HTML tags), má»i ngÆ°á»i cÃ³ thá»ƒ sá»­ dá»¥ng thÆ° viá»‡n **BeautifulSoup** Ä‘á»ƒ thá»±c hiá»‡n dá»… dÃ ng hÆ¡n.

DÆ°á»›i Ä‘Ã¢y lÃ  vÃ­ dá»¥ vá» cÃ¡ch dÃ¹ng.

````python
# Import thÆ° viá»‡n vÃ o
from bs4 import BeautifulSoup
````

````python
# Táº¡o cÃ¢u máº«u cÃ³ chá»©a tháº» HTML trong Ä‘Ã³
sentence = "Captain Teemo <br> on duty!" # <br> lÃ  tháº» trong HTML Ä‘á»ƒ xuá»‘ng hÃ ng
# Cho cÃ¢u máº«u qua BeautifulSoup Ä‘á»ƒ táº¡o Ä‘á»‘i tÆ°á»£ng soup Ä‘Ã£ Ä‘Æ°á»£c xá»­ lÃ½
soup = BeautifulSoup(sentence)
# Tiáº¿n hÃ nh chá»‰ trÃ­ch xuáº¥t má»—i dá»¯ liá»‡u vÄƒn báº£n, khÃ´ng bao gá»“m cÃ¡c tháº» HTML
sentence = soup.get_text()
print("CÃ¢u Ä‘Ã£ Ä‘Æ°á»£c loáº¡i bá» tháº» HTML: ", sentence)
````

Má»i ngÆ°á»i tháº¥y Ä‘áº¥y, tháº» <br> Ä‘Ã£ bá»‹ loáº¡i bá», hiá»‡n táº¡i khoáº£ng tráº¯ng giá»¯a 2 tá»« **Teemo** vÃ  **on** lÃ  2 kÃ­ tá»± khoáº£ng cÃ¡ch (space)

BÃ¢y giá» tá»¥i mÃ¬nh sáº½ tiáº¿n hÃ nh loáº¡i bá» tá»« dá»«ng cÅ©ng nhÆ° cÃ¡c dáº¥u cÃ¢u nha.

á» Ä‘Ã¢y, cÃ¡ch Ä‘Æ¡n giáº£n nháº¥t lÃ  má»i ngÆ°á»i láº·p qua cÃ¡c tá»« vÃ  kiá»ƒm tra xem tá»« Ä‘Ã³ cÃ³ trong danh sÃ¡ch tá»« dá»«ng khÃ´ng nha, náº¿u cÃ³ thÃ¬ tiáº¿n hÃ nh loáº¡i bá», cÃ²n khÃ´ng pháº£i thÃ¬ giá»¯ láº¡i.

Vá» pháº§n cÃ¡c dáº¥u cÃ¢u, chÃºng ta cÃ³ thá»ƒ sá»­ dá»¥ng phÆ°Æ¡ng thá»©c translate Ä‘á»ƒ tiáº¿n hÃ nh loáº¡i bá». Má»i ngÆ°á»i cÃ³ thá»ƒ sá»­ dá»¥ng **string.punctuation** Ä‘á»ƒ láº¥y danh sÃ¡ch cÃ¡c kÃ­ tá»± Ä‘áº·c biá»‡t vÃ  dáº¥u cÃ¢u nha.

````python
import string
# In ra danh sÃ¡ch cÃ¡c dáº¥u cÃ¢u vÃ  kÃ­ tá»± Ä‘áº·c biá»‡t.
print(string.punctuation)
````

````python
# Táº¡o báº£ng mapping Ä‘á»ƒ loáº¡i bá» kÃ­ tá»± Ä‘áº·c biá»‡t
table = str.maketrans('', '', string.punctuation) # Chuyá»ƒn cÃ¡c kÃ­ tá»± Ä‘áº·c biá»‡t thÃ nh kÃ­ tá»± rá»—ng
````

````python
# Láº¥y danh sÃ¡ch tá»« dá»«ng, á»Ÿ Ä‘Ã¢y mÃ¬nh sá»­ dá»¥ng thÆ° viá»‡n nltk Ä‘á»ƒ láº¥y tá»« dá»«ng trong tiáº¿ng Anh nha
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
stopwords = set(stopwords.words('english'))
print("Danh sÃ¡ch tá»« dá»«ng: ", stopwords)
````

````python
# Chuyá»ƒn danh sÃ¡ch tá»« dá»«ng sang dáº¡ng list
stopwords = list(stopwords)
````

````python
# Chia thÃ nh danh sÃ¡ch cÃ¡c tá»« ra
words = sentence.split()
print("Danh sÃ¡ch cÃ¡c tá»«: ", words)
# Chuá»—i káº¿t quáº£
filtered_sentence = ""
for word in words:
  word = word.translate(table)    # Loáº¡i bá» cÃ¡c kÃ­ tá»± Ä‘áº·c biá»‡t
  if word not in stopwords:       # Kiá»ƒm tra cÃ³ thuá»™c tá»« dá»«ng khÃ´ng Ä‘á»ƒ bá» qua.
    filtered_sentence = filtered_sentence + word + " "
print("CÃ¢u Ä‘Ã£ Ä‘Æ°á»£c tiá»n xá»­ lÃ½: ", filtered_sentence)
````
